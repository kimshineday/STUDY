### bool
bool, boolean : 참 또는 거짓, true 또는 false, 1 또는 0

bool의 쇼타임 시작
주의할점 True x true False x false

**부등호**
3 > 1

```python
# bool의 쇼타임 시작
# 주의할점 True x true False x false
# 부등호

3 > 1 # 'True'

3 == 1 # 'False' "=" 값을 할당할때 사용하는 기호. 같다는 부등호 표기 시에는 두개 사용

3 == 3 # 'True'

# 같지 않다
20 != 5 # 'True'

20 != 20 # 'False'

# 크거나 같다
5 >= 5 #True
5 >= 4 #True
5 <= 4 #False
```

```python
# 문자열 비교
'python' == 'Python' #False
'Python' == 'Python' #True
'python' != 'Python' #True
```

### 중요

```python
# ==, is 같은 기능을 한다

3 is 3 #True
3 == 3 #True

3.0 is 3 #False, 객체의 타입을 비교
3.0 == 3 #True, 값을 비교


print(id(3.0))
print(id(3))
#자료형이 생성되면서 메모리 주소가 생김. (고유 시리얼 넘버 같은?)
#출력 값이 다름. is를 사용할때 문제가 생길 수 있다.
```

### 본격적인 논리연산자

```python
# 현실세계에 있는 규칙들을 만든다.
# True : 1 / False : 0
True and True # 1 * 1 = 1
True and False # 1 * 0 = 0
False and True # 0 * 1 = 0
False and False # 0 * 0 = 0

#and는 두가지 값을 비교해서 동일할때 동일한 값을 출력
#두개의 값이 다를때 False 출력
```

```python
# True : 1 / False : 0
True or True # 1 + 1 = 1
True or False # 1 + 0 = 1
False or True # 0 + 1 = 1
False or False # 0 + 0 = 0

#or에서는 True에 힘이 실린다.
```

```python
#not
not True #False
not False #True
# 반대의 결과 값을 출력
```

```python
# 점점 복잡해지는 조건들!!!

10 == 12 or 6 != 2 # f or t = t
2 > 5 and 7 == 7 # f and t = f
not 14 <= 15 # f
not 7 is 7.0 # t
```

```python
# 문자열 왜 True?
bool("힘들다")
```

```python
# 중요한 내용
# 단락 평가
print(False and True) #False
print(False and '힘들어요') #False, 단락평가로 False and 까지만 인식을 하고 뒤에 값을 보지 않음

print(True and False) # 뒤에 값까지 비교 후 False 출력

print(True or '힘들어요') # 앞자리 값이 True, 뒤에 내용 상관 없이 True
print(True and '힘들어요') # 뒤에 나오는 값까지 확인하기 때문에 뒤에 값 출력

```
