```python
# 시퀀스 자료형에 다 지원되는 메소드
번호1, 번호2, 번호3, 번호4 = input().split(',')

번호1 = int(번호1)
번호2 = int(번호2)
번호3 = int(번호3)
번호4 = int(번호4)

print(type(번호1), 번호2, 번호3, 번호4)

#234234,56656,3444,5667
#<class 'int'> 56656 3444 5667
```

```python
#map()
번호1, 번호2, 번호3, 번호4 = map(int, input().split(','))
# 괄호 안에 적용시킬 값들을 넣는다. 맵 안에 쌓이게 됨
print(type(번호1), 번호2, 번호3, 번호4)

#123,3454,234,666
#<class 'int'> 3454 234 666
```

```python
# 레거시코드 : 누군가 남겨두고 간 코드
# 리펙토링 : 결과의 변경 없이 코드의 구조를 재조정 하는 것.
# input 사용

# 변수에 값을 할당
redbean = '붕어빵(팥)'
redbean__price = 2000
cream = '붕어빵(슈크림)'
cream__price = 2500
japchae = '붕어빵(잡채)'
japchae__price = 3000

# 주문 개수
#order2, order3 = 2, 3
#total = order2 + order3

# 주문서
redbean_, cream_, japchae_ = map(int, input("붕어빵 '팥', '슈크림', '잡채'의 개수를 쉼표로 구분지어 순서대로 입력해주세요. 원하지 않은 맛은 0  ").split(','))


# 총 결제 금액 산출
total__price = (redbean__price * redbean_) + (cream__price * cream_) + (japchae__price * japchae_)


# 출력
print(f'주문이 완료되었습니다. {redbean} {redbean_}개, {cream} {cream_}개, {japchae} {japchae_}개 총 {redbean_ + cream_ + japchae_}개 결제 금액은 {total__price}원 입니다.')

#붕어빵 '팥', '슈크림', '잡채'의 개수를 쉼표로 구분지어 순서대로 입력해주세요. 원하지 않은 맛은 0  2,5,7
#주문이 완료되었습니다. 붕어빵(팥) 2개, 붕어빵(슈크림) 5개, 붕어빵(잡채) 7개 총 14개 결제 금액은 37500원 입니다.
```
