### 변수란? (기본중에 기본, 근본 중에 근본, ....)
 #### 파이썬에서의 변수란..?
하나의 값을 저장할 수 있는 공간

List Dict 함수 .. 넣을 수 있다.

하나의 변수에 하나의 값을 넣을 때

변수이름 = 값

두개의 변수에 두개의 변수의 값을 넣을 때

변수이름1, 변수이름2 = 값1, 값2


```python
# 하나의 값을 변수에 넣을 때
num = 1
print(num)

# 두개의 값을 변수에 넣을 때
num_1, num_2 = 1, 2
print(num_1)
print(num_2)
print(num_1, num_2)
```

### 변수에 값을 넣는 것 -> 할당

단어, 용어에 대해 익숙해지는 것

sum = 1 + 2 -> 1+2의 계산된 결과를 sum이라는 변수에 할당한다.  
sum 변수명이고 '=' 같다는 의미가 아니라 할당(assignment)이라고 한다.  

'=' : 할당 연산자  
'+' = 산술 연산자

+= 산술 후 바로 할당하는 문법  
/=, -=, *= 도 가능

```python
# 변수를 사용하면서 사용 가능한 문법

sum = 1 + 2
sum = sum + 3
sum = sum + 4
sum = sum + 5

print(sum)

sum2 = 1 + 2
sum2 += 3
sum2 += 4
sum2 += 5

print(sum2)
```

```python
# 변수에 어떤값도 넣지 않고 선언하는 방법

nothing = None
print(nothing)
```

### 변수를 만드는 규칙
영문 문자와 숫자를 사욭할 수 있다.  
대소문자를 구분합다.  
첫 글자는 대문자로 쓰는 걸 지향한다.  
문자부터 시작해야하며, 숫자부터 시작하면 안된다.  
_(밑줄 문자)로 시작할 수 있다.
특수문자(+, -. *, /, $, @, &, % 등)는 사용할 수 없다.  
파이썬의 키워드(if, for, while, and, or 등)는 사용할 수 없다.


```python
# 대소문자를 구분하는 변수
BIG = '크다'
big = '작다'

print(BIG, big)
```

```python
# 붕어빵 철이 돌아왔죠
# 붕어빵(팥) : 2000원, 붕어빵(슈크림) : 2500원, 붕어빵(잡채) : 3000원
# 무인 주문기계를 만들어주세요
# 붕어빵(잡채) 3개, 붕어빵(슈크림) 2개를 주문하면
# "주문이 완료되었습니다. 붕어빵(슈크림)2개), 붕어빵(잡체) 3개 총 5개 결제 금액은 14000원 입니다."
# 출력되도록 코드를 작성해주세요.
# 스네이크 케이스에 언더바 2개 / print()문의 포메팅 f-string

# 변수에 값을 할당
redbean = '붕어빵(팥)'
redbean__price = 2000
cream = '붕어빵(슈크림)'
cream__price = 2500
japchae = '붕어빵(잡채)'
japchae__price = 3000

# 주문 개수
order2, order3 = 2, 3
total = order2 + order3

# 총 결제 금액 산출
total__price = cream__price * order2 + japchae__price * order3

# 출력
print(f'주문이 완료되었습니다. {cream} {order2}개, {japchae} {order3}개 총 {total}개 결제 금액은 {total__price}원 입니다.')
```
