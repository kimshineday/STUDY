### 슬라이싱
문자열을 잘라내거나 부분적으로 추출할때 사용한다.

퀴즈 = '차 4대가 고장나면? 카포에라'

print(퀴즈[12:16])

마지막은 항상 +1이된 값을 넣어줘야한다.

```python
퀴즈 = '차 4대가 고장나면? 카포에라'

print(퀴즈[12:16])
print(퀴즈[12:15])
```

```python
퀴즈[15] # 위치 뽑아서 하기 때문에 슬라이싱 기능을 이용할때만 +1 값 적용
퀴즈[-1] # 바로 끝에 있는 요소를 가져올 수 있음
```

```python
생년월일 = '2000.12.25' # >> 2000 12 25로 출력

print(생년월일)

연 = 생년월일[0:4]
월 = 생년월일[5:7]
일 = 생년월일[8:11]

print(f'{연} {월} {일}')
```

```python
숫자 = '1234567890'

#전체 출력
print(숫자[::])
print(숫자[:])

#반대로 출력하기
print(숫자[::-1])

#첫 번재 인덱스 기준으로 2칸씩 띄어쓰기 출력하기
print(숫자[::2])

# 2번째 자리부터 5번째 자리까지 반대로 출력하기
print(숫자[6:2:-1])
# 잘못 입력시에 출력이 안됨
```

퀴즈 = '스님이 공중에 뜬다를 4글자로 말하면? 어중이 떠중이'를 정답만 출력하기

```python
퀴즈 = '스님이 공중에 뜬다를 4글자로 말하면? 어중이 떠중이'
# 변수 값 할당
네글자 = len(퀴즈)

# 방법1, 글자 위치 값 찾기 및 출력
print(퀴즈[네글자-7:네글자+1])

# 방법2, 글자 위치 찾기
a = 네글자 - 7
b = 네글자 + 1

# 출력
print(퀴즈[a:b])
```
