### 복습
### 시퀀스 자료형
* 문자열 :
        * 시퀀스형에서 사용할 수 있는 메소드(count(), index(), split() 등..)
        * 승라이싱 ([:] [::])
        * 요소들로 이루어져 가능했던 기능들 (요소 합치기, 요소 찾기 등등)
* list :
        * 시퀀스형에서 사용할 수 있는 메소드 (count(), index(), split() 등..)
        * 슬라이싱 ([:] [::])
        * 요소들로 이루어져 가능했던 기능들 (요소 합치기, 요소 찾기 등등)
        * list 지원하는 다양한 메소드(append(), insert(), del, sort(), reverse())



### tuple :
        * 리스트처럼 요소를 일렬로 저장하지만 안에 저장된 요소를 변경, 추가, 삭제 할 수 없다.
        * 값을 추가, 삭제, 변경할 수 없기 때문에 사용하지 못하는 메소드가 많다.
        * tuple의 생성방법은 tuple = 값, 값 또는 tuple = (값, 값)
        * 그리고 엄청난 비밀이 있다.
        * 튜플은 소괄호


```python
# 리스트 [] 튜플 ()
튜플 = (1, '이', 3.0)
print(type(튜플))
print(튜플)

리스트 = [1, '이', 3]
print(type(리스트))
print(리스트)
```

```python
튜플2 = 1, '이', 3 # () 패킹을 안해도 튜플로 인식
print(type(튜플2))
print(튜플2)

튜플_솔로 = (1,) # 콤마를 안쓰면 수식으로 인식함
print(type(튜플_솔로))
print(튜플_솔로)
```

```python
# 값 추가, 삭제, 변경불가... 확인
print(튜플)
print(dir(튜플))

튜플_리스트화 = list(튜플)
print(type(튜플_리스트화))
```

```python
# 확인
튜플.append('두리안')
print(튜플)
```

```python
# 번외
수강생 = '중꺽그마'

수강생_list = list(수강생)
수강생_tuple = tuple(수강생)

print(수강생_list)
print(수강생_tuple)

수강생_list[1] = "뿌" # 튜플은 쓰지 못하는 기능
print(수강생_list)

# join함수 공부, 알고 있으면 코딩테스트에 좋다
취뽀_수강생 = ''.join(수강생_list) # 앞에 오는 것을 제거?
print(취뽀_수강생)

```

```python
# 공통점
print(튜플)

# 길이를 구할 수 있다.
len(튜플)
# 곱하기 연산으로 요소를 복사할 수 있다.
튜플 * 3
# 튜플은 요소들끼리 더할 수 있다.
튜플 + 튜플
# 튜플은 인덱싱이 가능하다.
튜플[2]
# 튜플의 인덱스
튜플.index(3.0)
# 튜플 요소 개수
튜플.count(1)
# 튜플도 슬라이싱 가능
튜플[1:3]
```

## tuple 사용이유

1. 넣기, 삭제, 바꾸기가 안된다  
여러명이 코딩을 하는 경우 변경되면 안되는 변수를 건드리는 경우  
또는 값이 변경되면 안되는 중요한 데이터를 넣어서 보안적으로 관리하기 좋다.
(은행계좌, 비밀번호 ...)

2. list는 tuple보다 무겁다.  
단지 기능이 많아서 무거운 것이 아니라 C언어로 만들어진 파이썬은 C 컴파일로 인해서 변환.
동적배열을 지원한다.  

    파이썬은 동적 배열을 지원한다.  
- list = [1, 2, 3, 4] 생성해서 4개에 값을 모두 넣었다.
- 문제는 그 다음 값을 추가할 때 이미 공간이 다 찼기 때문에 추가할당을 진행해야함.
    추가 할당 시 메모리 크기는 4, 8, 16, 25 더블링을 해준다.

3. 리스트는 가변, 튜플은 불변이다.  
값이 고정 되어야 하거나 크기가 작을 경우 튜플을 권장.



* 가변객체와 불변객체
* 파이썬 리스트의 동적배열
